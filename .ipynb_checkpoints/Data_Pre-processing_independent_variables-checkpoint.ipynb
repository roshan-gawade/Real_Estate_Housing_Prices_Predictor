# Processing Independent Variables

# imputing null values of numerical variables

```python
numerical_columns = ['No of Bathrooms', 'Flat Area (in Sqft)', 'Lot Area (in Sqft)', 'Area of the House from Basement (in Sqft)'
                    , 'Longitude', 'Latitude', 'Living Area after Renovation (in Sqft)']
# not considering 'Zipcode' as it should be treated as cateogorical variable
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values = np.nan, strategy = 'median') # imputing with median

data[numerical_columns] = imputer.fit_transform(data[numerical_columns])
```

```python
data.info()
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent') # imputing by mode
data['Zipcode'] = imputer.fit_transform(data['Zipcode'].values.reshape(-1,1))
```

```python
data.info()
```

```python
data.to_csv('data_1.1.csv', index = False)
```
