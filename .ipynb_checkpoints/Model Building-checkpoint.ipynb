```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('data_1.2.csv') # reading pre-processed data

data.head()
```

# Splitting Data

```python
X = data.iloc[:, 1:31]
Y = data.iloc[:, 0]

X.shape, Y.shape
```

 # Feature Scaling

```python
# Standardisation

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
X = pd.DataFrame(data = X, columns = data.drop(columns = 'Sale_Price').columns) # converting back to dataframe

X.head()
```

# Checking and Treating Multicollinearity

```python
X.corr()
```

```python
X_corr = X.corr()
X_multi_coll = [[str(i), str(j)] for i in X_corr.columns for j in X_corr.columns if (X_corr.loc[i, j] > abs(0.5)) & (i!=j)]

X_multi_coll, len(X_multi_coll) # i.e. 32/2 = 16 multicollineared pairs
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

VIF = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index = X.columns)

VIF # 7 variables showing multicollinearity i.e. VIF > 5 as it's calculated independently on variable
    # therefore it'll change as variable's dropped
```

```python
# remove multicollinearity using VIF (Variance Inflation Factor)

def mc_remover(data): # function to remove variables showing multicollinearity
    vif = pd.Series([variance_inflation_factor(data.values, i) for i in range(data.shape[1])], index = data.columns)
    
    if vif.max() > 5:
        print(vif[vif == vif.max()].index[0], 'has been removed')
        data = data.drop(columns = [vif[vif == vif.max()].index[0]])
        return data
    else:
        print("No more Multicollinearity present...!")
        return data
    
```

```python
for i in range(7):
    X = mc_remover(X)
    
X.head()
```

```python
# checking multicollinearity for remaining 28 columns

VIF = pd.Series([variance_inflation_factor(X.values, i) for i in range(X.shape[1])], index = X.columns)

VIF # no multicollinearity present
```

# Linear Regression Model

```python
# train-test split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)

x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# performing linear regression

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(normalize = True)
lin_reg.fit(x_train, y_train)
```

```python
lin_reg.score(x_test, y_test) # model got r-square score as below
```

```python
# calculating root mean squared error and mean absolute error
from sklearn.metrics import mean_squared_error, mean_absolute_error

predictions = lin_reg.predict(x_test)

lin_mse = mean_squared_error(y_test, predictions)
lin_rmse = np.sqrt(lin_mse)

lin_mae = mean_absolute_error(y_test, predictions)

print('rmse:', lin_rmse, 'mae:', lin_mae )
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor(n_estimators = 100, random_state = 42)
forest_reg.fit(x_train, y_train)
```

```python
forest_reg.score(x_test, y_test) # model got r-square score as below
```

```python
predictions = forest_reg.predict(x_test)

from sklearn.metrics import mean_squared_error, mean_absolute_error

forest_mse = mean_squared_error(y_test, predictions)
forest_rmse = np.sqrt(forest_mse)

forest_mae = mean_absolute_error(y_test, predictions)

print('rmse:', forest_rmse, 'mae:', forest_mae)
```
